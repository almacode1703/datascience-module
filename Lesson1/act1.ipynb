```python
import numpy as np

student_datatype = [("name", "U20"), ("class", "int16"), ("weight", "float")]
student_data = [("Vijay", "12", "65.5"),("Amit", "11", "71.5"),("Bharath", "10", "60.5")]

student_record = np.array(student_data, dtype=student_datatype)
print(student_record)
```

```python
print(f"Original Array : {student_record}")

sorted_weight = np.sort(student_record, order="weight")
sorted_name = np.sort(student_record, order="name")
sorted_data = np.sort(student_record, order="class")
print(f"After Sort by Weight : {sorted_weight}")
print(f"After Sort by Weight : {sorted_name}")
print(f"After Sort by class : {sorted_data}")
```
