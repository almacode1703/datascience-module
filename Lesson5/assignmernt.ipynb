```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Iris_Dataset.csv", encoding='latin1')

# Check columns (important)
print(df.columns)

# Optional: rename columns if needed
# df.columns = df.columns.str.strip()

# 1. Bar Plot
sns.barplot(data=df, x="Species", y="SepalLengthCm")
plt.title("Bar Plot: Species vs Sepal Length")
plt.show()

# 2. Count Plot
sns.countplot(data=df, x="Species")
plt.title("Count of Different Species")
plt.show()

# 3. Box Plot
sns.boxplot(data=df, x="Species", y="SepalWidthCm")
plt.title("Box Plot: Species vs Sepal Width")
plt.show()

# 4. Swarm Plot
sns.swarmplot(data=df, x="Species", y="SepalWidthCm")
plt.title("Swarm Plot: Species vs Sepal Width")
plt.show()

# 5. Distribution Plot
sns.histplot(df["SepalWidthCm"], kde=True)
plt.title("Distribution of Sepal Width")
plt.show()

# Joint Plot
sns.jointplot(data=df, x="SepalWidthCm", y="SepalLengthCm", kind="scatter")
plt.show()

# 6. Pair Plot
sns.pairplot(df, hue="Species")
plt.show()
```
