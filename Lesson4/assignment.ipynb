```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Tips Dataset.csv')
df.head(5)
```

```python
# Load dataset
tips = sns.load_dataset("tips")

plt.figure()
sns.histplot(data=tips, x="total_bill", kde=True)

plt.title("Histogram of Total Bill with KDE")
plt.show()
```
