```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv("IMDB.csv", encoding='latin1')

# 2. First 3 and last 3 rows
print("First 3 rows:\n", df.head(3))
print("\nLast 3 rows:\n", df.tail(3))

# 3. Detailed information
print("\nDataset Info:")
print(df.info())

# 4. Check null values
print("\nNull values:\n", df.isnull().sum())

# 5. Subset rows 41 to 75
subset = df.iloc[40:75]
print("\nSubset (41 to 75):\n", subset)

# 6. Movie with highest number of votes
max_votes_movie = df.loc[df['No_of_Votes'].idxmax()]
print("\nMovie with highest votes:\n", max_votes_movie)

# 7. Boxplot (IMDB Rating & Runtime)
sns.boxplot(data=df[['IMDB_Rating', 'Runtime']])
plt.title("Boxplot of IMDB Rating and Runtime")
plt.show()

# 8. Relationship (scatter plot using matplotlib)
plt.scatter(df['Runtime'], df['IMDB_Rating'])
plt.xlabel("Runtime")
plt.ylabel("IMDB Rating")
plt.title("Runtime vs IMDB Rating")
plt.show()

# 9. Distribution plots
sns.histplot(df['IMDB_Rating'], kde=True)
plt.title("Distribution of IMDB Rating")
plt.show()

sns.histplot(df['Runtime'], kde=True)
plt.title("Distribution of Runtime")
plt.show()

# 10. Count plot of Rating
sns.countplot(data=df, x="Rating")
plt.xticks(rotation=45)
plt.title("Count of Movies by Rating")
plt.show()
```
