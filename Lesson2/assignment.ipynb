```python
import pandas as pd
import numpy as np

data = {
    "Id": [1, 2, 3, 4],
    "Name": ["Pankaj", "Meghna", "David", "Lisa"],
    "Role": ["CEO", np.nan, np.nan, np.nan],
    "Salary": [100, 200, np.nan, np.nan]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print("First 2 rows:")
print(df.head(2))

print("\nLast 2 rows:")
print(df.tail(2))
```

```python
print("\nTotal null values:")
print(df.isnull().sum().sum())
```

```python
print("\nDataFrame Info:")
df.info()
```

```python
new_df = df.dropna()

print("\nAfter dropping null values:")
print(new_df)
```
